```python
print("Q_4_ANS :- Bokeh server is a Bokeh library feature that allows you to create interactive plots and applications that can be updated in real-time. It lets you serve Bokeh plots and widgets as web pages and applications, so users can interact with them from a web browser.\n\nTo use Bokeh server, you can create a Python script that defines your Bokeh plot and its associated widgets, and then use the bokeh serve command to start the Bokeh server and serve your application.\n\nHere's an example of how to use Bokeh server to create an interactive scatter plot that allows the user to adjust the color and size of the plot's circles in real-time:")
```

```python
import bokeh.io
bokeh.io.output_notebook()
from bokeh.io import curdoc
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.layouts import column
from bokeh.models.widgets import Select, Slider

# Prepare the initial data
x_data = [1, 2, 3, 4, 5]
y_data = [6, 7, 2, 4, 5]
colors = ['red', 'blue', 'green', 'orange', 'purple']
sizes = [10, 20, 30, 40, 50]
source = ColumnDataSource(data=dict(x=x_data, y=y_data, color=colors, size=sizes))

# Create the Bokeh plot
plot = figure(title="My Interactive Scatter Plot", x_axis_label="X", y_axis_label="Y")
plot.circle(x='x', y='y', color='color', size='size', source=source)

# Define the widgets
color_select = Select(title="Color", options=colors, value=colors[0])
size_slider = Slider(title="Size", value=sizes[0], start=10, end=50, step=10)

# Define the callback functions
def update_color(attrname, old, new):
    source.data['color'] = [colors[int(i)] for i in source.data['color_index']]
def update_size(attrname, old, new):
    source.data['size'] = [size_slider.value] * len(source.data['x'])

# Connect the widgets to the callbacks
color_select.on_change('value', update_color)
size_slider.on_change('value', update_size)

# Arrange the widgets in a layout
controls = column(color_select, size_slider)
layout = column(plot, controls)

# Add the layout to the current document
curdoc().add_root(layout)
```

```python
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # Create Bokeh plot
    plot = figure(title='My Plot', x_axis_label='X', y_axis_label='Y')
    plot.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5])
    
    # Generate Bokeh plot components
    script, div = components(plot)
    
    # Render the template file with the Bokeh plot
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run(debug=True)
```
